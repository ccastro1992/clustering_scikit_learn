```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
X, y, c = make_blobs(500, cluster_std=0.8, centers=4, n_features=2, return_centers=True)
```

```python
df_blob = pd.DataFrame({
    'x1': X[:, 0],
    'x2': X[:, 1],
    'y': y
})
```

```python
df_centers = pd.DataFrame({
    'x1': c[:, 0],
    'x2': c[:, 1]
})
```

```python
sns.scatterplot(data=df_blob, x='x1', y='x2')
sns.scatterplot(data=df_centers, x='x1', y='x2', markers='X', s=100)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
df_cluster = kmeans.fit_predict(X)
df_blob['cluster'] = df_cluster
k_means_centers = kmeans.cluster_centers_

df_k_means_centers = pd.DataFrame({
    'x1': k_means_centers[:, 0],
    'x2': k_means_centers[:, 1],
})
```

```python
sns.scatterplot(data=df_blob, x='x1', y='x2', hue='cluster', palette='coolwarm')
sns.scatterplot(data=df_centers, x='x1', y='x2', marker='X', s=150, color='red')
sns.scatterplot(data=df_k_means_centers, x='x1', y='x2', marker='o', s=150, color='yellow')
plt.show()
```

```python
def vis_cluster(k):
    kmeans = KMeans(n_clusters=k)
    df_cluster = kmeans.fit_predict(X)
    df_blob['cluster'] = df_cluster
    k_means_centers = kmeans.cluster_centers_

    df_k_means_centers = pd.DataFrame({
        'x1': k_means_centers[:, 0],
        'x2': k_means_centers[:, 1],
    })

    sns.scatterplot(data=df_blob, x='x1', y='x2', hue='cluster', palette='coolwarm')
    sns.scatterplot(data=df_centers, x='x1', y='x2', marker='X', s=150, color='red')
    sns.scatterplot(data=df_k_means_centers, x='x1', y='x2', marker='o', s=150, color='yellow')
    plt.show()
```

```python
for _ in range(3, 7):
    print(_)
    vis_cluster(_)
```

```python
sum_of_squared_distances = []
K = range(2, 15)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    sum_of_squared_distances.append(kmeans.inertia_)

plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('K')
plt.ylabel('Inertia')
plt.show()
```

```python
silhouette_scores = []
K = range(2, 15)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    y = kmeans.predict(X)
    silhouette_scores.append(silhouette_score(X, y))

plt.plot(K, silhouette_scores, 'bx-')
plt.xlabel('K')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
vis_cluster(4)
```

```python
km = KMeans(n_clusters=4)
vis = SilhouetteVisualizer(km, colors='yellowbrick')
vis.fit(X)
vis.show()
```
