```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_countries_scaled = scaler.fit_transform(df_countries.drop('country', axis=1))

df_countries_scaled
```

#### Reduccion Dimensionalidad

```python
df_country_scaled = pd.DataFrame(df_countries_scaled, columns=df_countries.drop('country', axis=1).columns)
```

#### PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(df_country_scaled)
pca_data_scaled = pca.transform(df_country_scaled)
```

```python
var = pca.explained_variance_ratio_
print(var)
```

```python
cum_var = np.cumsum(np.round(var, decimals=4) * 100)
plt.figure(figsize=(15, 10))
plt.plot(cum_var, 'r-x')
plt.show()
```

```python
pca_data_standard = pd.DataFrame(pca_data_scaled)
pca_data_standard.drop([4,5,6,7,8], axis=1, inplace=True)

pca_data_standard
```

#### KMEANS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
sum_of_squared_distances = []
silhouette_scores = []
K = range(2, 15)
for _ in K:
    km = KMeans(n_clusters=_)
    y = km.fit_predict(pca_data_standard)
    sum_of_squared_distances.append(km.inertia_)
    silhouette_scores.append(silhouette_score(pca_data_standard, y))
```

```python
plt.figure(figsize=(8, 8))
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('K')
plt.ylabel('Silhouette coefficient')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.plot(K, silhouette_scores, 'rx-')
plt.xlabel('K')
plt.ylabel('Silhouette coefficient')
plt.show()
```

```python
# KMEANS con 4 clusters, que es el ideal segun grafica de siluetas
km = KMeans(n_clusters=4)
y = km.fit_predict(pca_data_standard)
print(silhouette_score(pca_data_standard, y))

df_countries['kmeans_pcs'] = y
```
